# MVTec data loader test

## Imports

```python
import mvtec
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import torchvision.transforms.functional as F
```

## Define image plotting function

```python
def show(img, PIL=True):
    if not(PIL):
        img = F.to_pil_image(img)
        
    plt.imshow(np.asarray(img))
```

## Toothbrush train data (i.e., 'good' images),
## resize=64, interpolation=3, transform=ToTensor

```python
IMSHAPE = 64
transform = transforms.Compose(
    [transforms.ToTensor()])

data = mvtec.MVTEC(root="./mvtec",
                   train=True,
                   transform=transform,
                   resize=IMSHAPE,
                   interpolation=3,
                   category='toothbrush')
```

### test __len__ function

```python
# see table 1:
# https://openaccess.thecvf.com/content_CVPR_2019/papers/Bergmann_MVTec_AD_--_A_Comprehensive_Real-World_Dataset_for_Unsupervised_Anomaly_CVPR_2019_paper.pdf
n_toothbrush = data.__len__()
assert n_toothbrush == 60
```

### Define random index

```python
np.random.seed(8)
```

```python
index = np.random.randint(0, n_toothbrush)
print('index: ', index)
```

```python
print('label: ', data[index][1])
show(data[index][0], PIL=False)
```

### test __getitem__ function

```python
img, target = data.__getitem__(index)
print('label: ', target)
show(img, PIL=False)
```

## Test pytorch data loader with pill test data (i.e., 'good' and 'defective' images),
## resize=128, interpolation=3, transform=ToTensor, no shuffle

```python
from torch.utils.data import DataLoader
```

```python
IMSHAPE = 128
transform = transforms.Compose(
    [transforms.ToTensor()])

data = mvtec.MVTEC(root="./mvtec",
                   train=False,
                   transform=transform,
                   resize=IMSHAPE,
                   interpolation=3,
                   category='pill')

data_loader = DataLoader(dataset=data, batch_size=20, shuffle=False)
```

```python
for batch in data_loader:
    images = batch[0]
    labels = batch[1]
    print('images shape: (n,C,H,W) ', images.shape)
    print('labels shape: (n,) ', labels.shape)
    print('labels: ', labels)
    print('first image in batch:')
    show(images[0], PIL=False)
    break
```
